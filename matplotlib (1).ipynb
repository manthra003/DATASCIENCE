```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Create one subplot
plt.subplot(1, 1, 1)

# Plot
plt.plot(x, y, marker='o', label="Line")

# Labels and title
plt.title("Simple Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Grid and legend
plt.grid(True)
plt.legend()

# Show
plt.show()


import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.array(["A", "B", "C", "D"])
y = np.array([10, 20, 15, 25])

# Bar graph
plt.bar(x, y, label="Marks")

# Title and labels
plt.title("Bar Graph Example")
plt.xlabel("Categories")
plt.ylabel("Values")

# Grid and legend
plt.grid(True)
plt.legend()

# Show
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Categories (purchase behaviour)
behaviour = np.array(["Online", "In-Store", "Subscription", "Impulse"])

# Values (price spent)
price = np.array([2000, 1500, 3000, 1200])

# Bar graph
plt.bar(behaviour, price, label="Purchase Amount")

# Title and labels
plt.title("Purchase Behaviour vs Price")
plt.xlabel("Behaviour Type")
plt.ylabel("Price (₹)")

# Grid and legend
plt.grid(True)
plt.legend()

# Show graph
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate continuous price data
price = np.random.normal(loc=2000, scale=500, size=100)

# Histogram
plt.hist(price, bins=10, edgecolor='black')

# Labels
plt.title("Histogram of Purchase Prices")
plt.xlabel("Price (₹)")
plt.ylabel("Frequency")

plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 2D array (6 rows → 6 graphs)
data = np.array([
    [1, 2, 3, 4],
    [2, 4, 6, 8],
    [5, 3, 2, 1],
    [4, 6, 8, 10],
    [7, 5, 3, 1],   # ← this is 5th row
    [2, 5, 7, 9]
])

x = np.array([1, 2, 3, 4])

# Select subplot 5
plt.subplot(2, 3, 5)

# Plot 5th row (index 4)
plt.plot(x, data[4], marker='o', label="Plot 5")

# Labels
plt.title("Subplot 5 (from 2D array)")
plt.xlabel("X")
plt.ylabel("Y")

# Grid and legend
plt.grid(True)
plt.legend()

plt.show()
```
