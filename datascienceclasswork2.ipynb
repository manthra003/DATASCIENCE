```python
import pandas as pd
import numpy as np

# Creating employee DataFrame
data = {
    "emp_id": [101, 102, 103, 104, 105],
    "name": ["Arun", "Meena", "John", "Sara", "Kiran"],
    "designation": ["Manager", np.nan, "Developer", "Analyst", np.nan],
    "yrs_of_experience": [5, 2, 7, 3, 4],
    "age": [30, 25, 35, 28, 32],
    "salary": [15000, 8000, 12000, 9000, 11000]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
df = df[df["salary"] <= 10000]

print("\nAfter dropping salary > 10000:")
print(df)
df["designation"].fillna("Not Specified", inplace=True)

print("\nAfter handling NaN in designation:")
print(df)
df["age"] = df["age"] + 3
df.rename(columns={"yrs_of_experience": "experience"}, inplace=True)

print("\nFinal Cleaned DataFrame:")
print(df)
```

```python
import pandas as pd

# Student Details DataFrame
students_data = {
    "student_id": [1, 2, 3, 4],
    "name": ["Arjun", "Priya", "Rahul", "Sneha"],
    "age": [20, 21, 19, 22],
    "department": ["CSE", "ECE", "IT", "CSE"]
}

students_df = pd.DataFrame(students_data)


# Marks & Specialization DataFrame
marks_data = {
    "student_id": [1, 2, 3, 5],   # Notice 5 is extra
    "maths": [85, 78, 92, 70],
    "physics": [80, 74, 89, 65],
    "specialization": ["AI", "Embedded", "Data Science", "Cyber"]
}

marks_df = pd.DataFrame(marks_data)


# -------------------------------
# 1️⃣ INNER JOIN
# -------------------------------
inner_merge = pd.merge(students_df, marks_df, on="student_id", how="inner")
print("INNER JOIN:")
print(inner_merge)


# -------------------------------
# 2️⃣ OUTER JOIN
# -------------------------------
outer_merge = pd.merge(students_df, marks_df, on="student_id", how="outer")
print("\nOUTER JOIN:")
print(outer_merge)


# -------------------------------
# 3️⃣ LEFT JOIN
# -------------------------------
left_merge = pd.merge(students_df, marks_df, on="student_id", how="left")
print("\nLEFT JOIN:")
print(left_merge)


# -------------------------------
# 4️⃣ RIGHT JOIN
# -------------------------------
right_merge = pd.merge(students_df, marks_df, on="student_id", how="right")
print("\nRIGHT JOIN:")
print(right_merge)


# -------------------------------
# 5️⃣ CONCAT
# -------------------------------

# Vertical Concatenation (Row-wise)
concat_vertical = pd.concat([students_df, marks_df], axis=0)
print("\nCONCAT (Vertical):")
print(concat_vertical)

# Horizontal Concatenation (Column-wise)
concat_horizontal = pd.concat([students_df, marks_df], axis=1)
print("\nCONCAT (Horizontal):")
print(concat_horizontal)
```

Create Sales Data

```python
import pandas as pd

sales_data = {
    "Month": ["Jan", "Feb", "Mar"],
    "Product_A": [10000, 12000, 9000],
    "Product_B": [15000, 13000, 11000],
    "Product_C": [8000, 9500, 7000]
}

sales_df = pd.DataFrame(sales_data)

print("Original Sales Data (Wide Format):")
print(sales_df)
```

Use melt() (Wide ➝ Long)

```python
melted_df = pd.melt(
    sales_df,
    id_vars=["Month"],          # Column to keep fixed
    var_name="Product",         # New column name for product
    value_name="Sales"          # New column name for sales value
)

print("\nMelted Data (Long Format):")
print(melted_df)
```

Use pivot() (Long ➝ Wide)

```python
pivot_df = melted_df.pivot(
    index="Month",
    columns="Product",
    values="Sales"
)

print("\nPivoted Data (Back to Wide Format):")
print(pivot_df)
```
