1.to create a numpy array with an example

```python
import numpy as np

# 1D array
a = np.array([10, 20, 30])
print("Array a:", a)
print("ndim of a:", a.ndim)

# 2D array
b = np.array([[1, 2, 3],
              [4, 5, 6]])
print("\nArray b:")
print(b)
print("ndim of b:", b.ndim)

# 3D array
c = np.array([[[1, 2], [3, 4]],
              [[5, 6], [7, 8]]])
print("\nArray c:")
print(c)
print("ndim of c:", c.ndim)
```

3.To generate a value of arrays from range from 12-15 with a step value of 3

```python
import numpy as np

# Create values from 15 to 30
arr = np.arange(15, 31)

# Reshape into a 2D array (4 rows, 4 columns)
arr_2d = arr.reshape(4, 4)

print("2D Array:")
print(arr_2d)

print("\nShape of the array:", arr_2d.shape)
print("Size of the array:", arr_2d.size)
```

7.Difference between np.eye and np.identity

ans

np.eye
Creates a matrix with 1s on the diagonal and 0s elsewhere.
It is more flexible because it can create square or rectangular matrices.

np.identity
Creates a square identity matrix only, with 1s on the main diagonal and 0s elsewhere.
It is simple and fixed in size.
